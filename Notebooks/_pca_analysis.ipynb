```python
import sys, os
sys.path.append(os.path.abspath(".."))

import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from src.utils import N_selector, Plotter

# Load processed data
X_train = pd.read_csv("../data/X_train_processed.csv")
X_test = pd.read_csv("../data/X_test_processed.csv")
y_train = pd.read_csv("../data/y_train.csv").squeeze()

n_selec = N_selector(X_train, 0.95)
n_opt = n_selec.calc_n()
pca_optimal = PCA(n_components= n_opt)
X_train_pca = pca_optimal.fit_transform(X_train)

pca_plot = Plotter(X_train_pca)
pca_plot.scatter(0,1,y_train)
pca_plot.label("First Principal component", "Second Principal Component", "PCA Result Scatter Plot")

plt.plot(n_selec.cumulative_variance)
pca_plot.label("Number of Components", "Cumulative Variance", "Cumulative Variance Plot")
```
